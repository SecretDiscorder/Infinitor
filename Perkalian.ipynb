```python
!pip install reportlab
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from reportlab.lib.units import inch
import sys

def create_multiplication_table_pdf(filename, row_start, row_end, col_start, col_end):
    c = canvas.Canvas(filename, pagesize=letter)
    width, height = letter

    # Define cell size
    cell_width = 40
    cell_height = 20
    rows_per_page = 10
    cols_per_page = 10

    # Initialize start and end for columns and rows
    col_current_start = col_start
    col_current_end = min(col_start + cols_per_page - 1, col_end)
    row_current_start = row_start
    row_current_end = min(row_start + rows_per_page - 1, row_end)

    page_number = 0

    # Calculate total number of cells to process
    total_cells = (row_end - row_start + 1) * (col_end - col_start + 1)
    cells_processed = 0

    def print_progress(cells_processed, total_cells):
        progress_percentage = (cells_processed / total_cells) * 100
        sys.stdout.write(f"\rProgress: {progress_percentage:.2f}%")
        sys.stdout.flush()

    while row_current_start <= row_end:
        # Start a new page and set positions
        c.showPage()
        c.setFont("Helvetica", 10)
        current_x = inch
        current_y = height - inch

        # Draw table header
        c.drawString(current_x, current_y, "X")
        for i in range(col_current_start, col_current_end + 1):
            c.drawString(current_x + (i - col_current_start + 1) * cell_width, current_y, str(i))

        current_y -= cell_height
        row_end_on_page = min(row_current_start + rows_per_page - 1, row_end)

        for row_number in range(row_current_start, row_end_on_page + 1):
            if current_y - cell_height < inch:
                # If there's not enough space for the next row, create a new page
                c.showPage()
                c.setFont("Helvetica", 10)
                current_x = inch
                current_y = height - inch

                # Draw table header again
                c.drawString(current_x, current_y, "X")
                for i in range(col_current_start, col_current_end + 1):
                    c.drawString(current_x + (i - col_current_start + 1) * cell_width, current_y, str(i))

                current_y -= cell_height

            # Draw row data
            c.drawString(current_x, current_y, str(row_number))
            for j in range(col_current_start, col_current_end + 1):
                c.drawString(current_x + (j - col_current_start + 1) * cell_width, current_y, str(row_number * j))
                cells_processed += 1
                print_progress(cells_processed, total_cells)

            current_y -= cell_height

        # Move to the next set of columns
        col_current_start = col_current_end + 1
        col_current_end = min(col_current_start + cols_per_page - 1, col_end)

        if col_current_start > col_end:
            # Move to the next set of rows
            row_current_start = row_end_on_page + 1
            row_current_end = min(row_current_start + rows_per_page - 1, row_end)
            # Reset column range for new set of rows
            col_current_start = col_start
            col_current_end = min(col_start + cols_per_page - 1, col_end)

    c.save()
    print("\nPDF generation complete!")

# Input from the user
row_start = int(input("Enter start number for rows: "))
row_end = int(input("Enter end number for rows: "))
col_start = int(input("Enter start number for columns: "))
col_end = int(input("Enter end number for columns: "))

create_multiplication_table_pdf("multiplication_table.pdf", row_start, row_end, col_start, col_end)
```
